```python
import os
import json
import pandas as pd

city = "qd"#22091


datapath = f"{os.getcwd()}\\{city}"
filename = os.listdir(datapath)
# 过滤只剩下json文件
jsonname = [i for i in filename if "json" in i]
jsonname.sort()


b = []
for i in jsonname:
    with open(datapath + "\\" + i, "r") as f:
        load_dict = json.load(f)

    # 取出该文件中的所有酒店id
    a = list(load_dict["item"].keys())

    for i in a:
        # 增加酒店id到字典中
        load_dict["item"][i]["id"] = i
        # 取出酒店的订单记录字典
        load_dict["item"][i]["开始记录时间"] = load_dict["开始记录时间"]
        load_dict["item"][i]["结束记录时间"] = load_dict["结束记录时间"]
        t = pd.DataFrame(load_dict["item"][i])
        b.append(t)

# 将字典组成DF 对象
pp = pd.concat(b)
listname = ["开始记录时间", "结束记录时间", "id", "hotel_name",
            "hotel_value", "hotel_price", "order_time"]

pp = pp[listname]
pp.drop_duplicates(keep="first",inplace=True)
pp.sort_values(["order_time"],inplace=True)
pp.reset_index(drop="index",inplace=True)
print(pp.shape)
pp.to_csv(f"{city}.csv",encoding="utf-8-sig")
```

```python
import os
import json
import pandas as pd
```

```python
datapath = os.getcwd()+"\\cd"
filename = os.listdir(datapath)
jsonname = [i for i in filename if "json" in i]
# jsonname
```

```python
# 取出文件名,即日期
daset=set()
for i in jsonname:
    r=i.split(".")[0][:8]
    daset.add(r)
daset=list(daset)
daset.sort()
```

```python
# 按天进行查询时间
fidata = []
for k in daset:
    fidata.append([i for i in jsonname if k in i])
# fidata
```

```python
x, y, z = [], [], []
# 创建一个字典
orderData = {"date": []}
for i in range(24):
    orderData[f"{i}h"] = []
index = 0
for uu in fidata:
    # 初始化
    orderData["date"].append("")
    for i in range(24):
        orderData[f"{i}h"] .append("")
    orderData["date"][index] = uu[0][:8]
    for k in range(24):
        orderData[f"{k}h"][index] = 0

    # 特定时间中文件有多少份
    for j in range(len(uu)):
        
        with open(datapath + "\\" + uu[j], "r") as f:
            load_dict = json.load(f)
            Stime = int(load_dict['开始记录时间'][11:13])
            Etime = int(load_dict['结束记录时间'][11:13])
            
            
            S=Stime
            E=Etime
            if j==0 and Stime==23:
                for i in range(Etime):
                    orderData[f"{i}h"][index] = 1
            else:
                orderData[f"{S}h"][index] = 1
                for i in range(1,(E-S)+1):
                    orderData[f"{S+i}h"][index] = 1            
            
    index+=1
```
